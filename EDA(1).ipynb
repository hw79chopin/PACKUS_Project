```python
def div_card(row):
    if pd.isnull(row):
        return '미입력' 
    
    if '삼성' in row:
        return '삼성'
    elif '신한' in row:
        return '신한'
    elif '현대' in row:
        return '현대'
    elif '비씨' in row:
        return '비씨'
    elif '국민' in row:
        return '국민'
    elif '농협' in row:
        return '농협'
    elif '롯데' in row:
        return '롯데'
    elif '하나' in row:
        return '하나'
    else:
        return '기타'
    
product_data['카드사(clean)'] = product_data['카드사'].apply(div_card)
product_data['카드사(clean)'].value_counts()
```

```python
def div_cate(row):
    if '무통장' in row:
        return '무통장'
    elif '카드' in row:
        return '카드결제'
    elif '휴대폰' in row:
        return '휴대폰결제'
    elif '실시간' in row:
        return '실시간계좌이체'
    else:
        return '기타'

product_data['결제수단(clean)'] = product_data['결제수단'].apply(div_cate)
product_data['결제수단(clean)'].value_counts()
```

```python
def div_cacul(row):
    list_month = [1,3,5,7,8,10,12]
    if row['배송완료월(dt)']== row['배송시작월(dt)']:
        day_e = row['배송완료일(day)']
        day_s = row['배송시작일(day)']
        return day_e - day_s
    
    elif row['배송완료월(dt)'] in list_month :
        day_e = row['배송완료일(day)']
        day_s = row['배송시작일(day)']
        return (31+(day_e)) - day_s
    else: 
        day_e = row['배송완료일(day)']
        day_s = row['배송시작일(day)']
        return (30+(day_e)) - day_s
    
product_data['배송기간'] = product_data[['배송완료일(dt)', '배송시작일(dt)',
       '배송완료월(dt)', '배송시작월(dt)',
      '배송완료일(day)','배송시작일(day)']].apply(div_cacul, axis=1)
```
